LINKED LIST

```python
class Node:
    def __init__(self,data,next):
        self.data=data
        self.next=next

class LinkedList:
    def __init__(self):
        self.head=None

    def insert_at_beginning(self,data):
        if self.head is None:
            d=Node(data,None)
            self.head=d
            return
        d=Node(data,self.head)
        self.head=d
    def print(self):
        itr=self.head
        if itr is None:
            print("Nothing in this Linked List")
            return
        r=''
        while itr:
            r+=str(itr.data)+"-->"
            itr=itr.next
        print(r)
    def count(self):
        n=0
        
        itr=self.head
        if itr is None:
            print(0)
            return 0
        while itr:
            itr=itr.next
            n+=1
       
        return n
    def insert_at_end(self,data):
        if self.head is None:
            self.insert_at_beginning(data)
            return
        itr=self.head
        c=0
        count=self.count()
        while itr:
            c+=1
            if c==count:
                d=Node(data,None)
                itr.next=d
                break
            itr=itr.next
    


    def insert_values(self,list):
        self.head=None
        for data in list:
            self.insert_at_end(data)

    def insert_at_index(self,index,data):
        if index<0 or index>=self.count():
            raise Exception("Invalid index da loosuh")
        if index==0:
            self.insert_at_beginning(data)
            return
        itr=self.head
        n=0
        while itr:
            
            if n==index-1:
                d=Node(data,itr.next)
                itr.next=d
                break
            itr=itr.next
            n+=1
    def delete_at_index(self,index):
        if index<0 or index>=self.count():
            raise Exception("Invalid index da loosuh")
        if index==0:
            self.head=self.head.next
            return
        n=0
        itr=self.head
        while itr:
            
            if n==index-1:
                itr.next  =itr.next.next
                break
            itr=itr.next
            n+=1

    def insert_after_value(self,value,data):
        itr=self.head
        while itr:
            if itr.data==value:
                d=Node(data,itr.next)
                itr.next=d
                return
            itr=itr.next

        raise Exception("not found (spelling olunga poodu)")
    def remove_by_value(self,value):
        itr=self.head
        if self.head.data==value:
            self.head=self.head.next
            return
        while itr:
            
            ne=itr.next
            if ne.data==value:
                itr.next=itr.next.next
                return
            itr=itr.next
```

Doubly Linkedlist

```python
class Nodes:
    def __init__(self,data,next,prev):
        self.data=data
        self.next=next
        self.prev=prev

class DoublyLL:
    def __init__(self):
        self.head=None
    def insert_at_beginning(self,data):
        if self.head is None:
            d=Nodes(data,None,None)
            self.head=d
            return
        d=Nodes(data,self.head,None)
        self.head=d
    def insert_at_end(self,data):
        if self.head is None:
            self.insert_at_beginning(data)
            return
        itr=self.head
        c=0
        count=self.count()
        while itr:
            c+=1
            if c==count:
                d=Nodes(data,None,itr)
                itr.next=d
                break
            itr=itr.next
    
    def count(self):
        n=0
        
        itr=self.head
        if itr is None:
            print(0)
            return 0
        while itr:
            itr=itr.next
            n+=1
        return n

    def insert_values(self,list):
        self.head=None
        for data in list:
            self.insert_at_end(data)
    
    def print_forward(self):
        r=''
        itr=self.head
        while itr:
            r+=str(itr.data)+"-->"
            itr=itr.next
        print(r)
    def print_backward(self):    
        r=''
        itr=self.head
        n=0
        while itr:
            if n == self.count()-1:
                rev=itr
                while rev:
                    r+="<--"+str(rev.data)
                    rev=rev.prev
            itr=itr.next
            n+=1
        print(r)
```

HASH MAP

```python
class HashMap:
    def __init__(self):
        self.MAX=100
        self.arr=[[] for i in range(self.MAX)]

    def hash(self,key):
        n=0
        for data in key:
            n+=ord(data)
        return n%self.MAX
    def __setitem__(self,key,value):
        n=self.hash(key)
        for pair in self.arr[n]:
            if pair[0]==key:
                pair[1]= value
                return
        self.arr[n].append([key,value])
           
        
    def __getitem__(self,key):
        n=self.hash(key)
        for pair in self.arr[n]:
            if pair[0]==key:
                return pair[1]
        raise Exception("key error da loosuh")
    
    def __delitem__(self,key):
        n=self.hash(key)
        for i,pair in enumerate(self.arr[n]):
            if pair[0]==key:
                del self.arr[n][i]
                return
        raise Exception("key error da loosuh")
```

STACK

```python
from collections import deque

class Stack:
    def __init__(self):
        self.container=deque()
    def push(self,value):
        self.container.append(value)
    def peek(self):
        return self.container[-1]
    def pop(self):
        n=self.container[-1]
        del self.container[-1]
        return n
    def print(self):
        print(self.container)

    def reverse_string(self,string):
        for char in string:
            self.push(char)
        return "".join(self.pop() for _ in range(len(self.container)))
    
```

```python
s=Stack()
s.reverse_string("kailainathan")
```

```python
def is_balanced(string):
    s=Stack()
    pair={']':'[','}':'{',')':'('}
    for char in string:
        if char in '{([':
            s.push(char)
        else:
            if not s.container:
                return False
            if s.pop()!=pair[char]:
                return False
    return len(s.container)==0
```

Queue

```python
from collections import deque
class Queue:
    def __init__(self):
        self.buffer=deque()
    def enque(self,value):
        self.buffer.appendleft(value)
    def deque(self):
        if len(self.buffer)==0:
            print("Queue is empty")
            return
        return self.buffer.pop()
    def is_empty(self):
        return len(self.buffer)==0
    def print(self):
        return self.buffer
    def front(self):
        if self.is_empty():
            return None
        return self.buffer[-1]
```

```python
import time

food = Queue()

def place_orders(orders):
    for order in orders:
        print("Placing order:", order)
        food.enque(order)
        time.sleep(0.5)
def server(total_orders):
    time.sleep(1)
    served = 0
    while served < total_orders:
        if food.is_empty():
            time.sleep(0.5)
            continue

        order = food.deque()
        print("Serving order:", order)
        served += 1
        time.sleep(2)

    print("All orders served. Server stopped.")
```

```python
orders = ["pizza", "burger", "briyani", "rice", "gril"]

import threading

t1 = threading.Thread(target=place_orders, args=(orders,))
t2 = threading.Thread(target=server, args=(len(orders),))

t1.start()
t2.start()
```

Binary search tree

```python
class BST:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
    def add(self,d):
        if self.data == d:
            return
        node=BST(d)
        if d<self.data:
            if self.left:
                self.left.add(d)
            else:
                self.left=node
        else:
            if self.right:
                self.right.add(d)
            else:
                self.right=node

    def inorder(self):
        elements=[]
        if self.left:
            elements+=self.left.inorder()
        elements.append(self.data)
        if self.right:
            elements+=self.right.inorder()
        return elements
    def search(self,val):
        if self.data== val:
            return True
        if val<self.data:
            if self.left:
                return self.left.search(val)
          
            return False
        else:
            if self.right:
                return self.right.search(val)
          
            return False
        
    def find_min(self):
        
        if self.left:
            return self.left.find_min()
        return self.data
    def find_max(self):
        if self.right:
            return self.right.find_max()
        return self.data
    def calculate_sum(self):
        return sum(self.inorder())
    def preorder(self):
            elements=[]
            elements.append(self.data)
            if self.left:
                elements+=self.left.preorder()
           
            if self.right:
                elements+=self.right.preorder()
            return elements
    def postorder(self):
            elements=[]
            if self.left:
                elements+=self.left.postorder()
            
            if self.right:
                elements+=self.right.postorder()
            elements.append(self.data)
            return elements
    def delete(self, val):
        if val < self.data:
            if self.left:
                self.left = self.left.delete(val)
        elif val > self.data:
            if self.right:
                self.right = self.right.delete(val)
        else:
          
            if self.left is None:  #case 1&2
                return self.right
            if self.right is None:
                return self.left

            max_val = self.left.find_max() #case 3
            self.data = max_val
            self.left = self.left.delete(max_val)

        return self


    



def builder(elements):
    root=BST(elements[0])
    for i in range(1,len(elements)):
        root.add(elements[i])
    return root
    


```

Binary Search

```python
def binary_search(arr,target):
    l=0
    r=len(arr)-1
    while l<=r:
        mid=(l+r)//2
        if arr[mid]==target:
            return mid
        elif arr[mid]>target:
            r=mid-1
        else:
            l=mid+1
def bs_rec(arr,target,l,r):
    while l<=r:
        mid=(l+r)//2
        if arr[mid]==target:
            return mid
        elif arr[mid]>target:
            r=mid-1
        else:
            l=mid+1
    return bs_rec(arr,target,l,r)


arr=[1,2,3,4,5,6,7,8]
print("basic bs:",binary_search(arr,6))
print("rec approach:",bs_rec(arr,6,0,len(arr)))
```

Bubble Sort

```python
def bubble_sort(arr):
    for i in range(len(arr)-1):
        swap=False

        for j in range(len(arr)-1-i):
            if arr[j]>arr[j+1]:
                temp=arr[j]
                arr[j]=arr[j+1]
                arr[j+1]=temp
                swap=True
        if swap==False:
            break

    return arr

l=[2,1,5,3,9,4,1,3,7,4,7,3]
print(bubble_sort(l))
```

```python
def bubble_sort_dict(arr,key):
    for i in range(len(arr)-1):
        swap=False

        for j in range(len(arr)-1-i):
            if arr[j][key]>arr[j+1][key]:
                temp=arr[j]
                arr[j]=arr[j+1]
                arr[j+1]=temp
                swap=True
        if swap==False:
            break

    return arr


elements = [
        { 'name': 'kathy',  'transaction_amount': 200,  'device': 'vivo'},
        { 'name': 'dhaval', 'transaction_amount': 400,  'device': 'google pixel'},
        { 'name': 'aamir',  'transaction_amount': 800,  'device': 'iphone-8'},
        { 'name': 'mona',   'transaction_amount': 1000, 'device': 'iphone-10'},
    ]

bubble_sort_dict(elements,'name')
```

Selection Sort

```python
def selection_sort(arr):
    n=len(arr)
    for i in range(n-1):
        minindex=i
        for j in range(i+1,n):
           if arr[minindex]>arr[j]:
                    minindex=j
        arr[i],arr[minindex]=arr[minindex],arr[i]
    return arr

arr=[1,3,2,78,9,2,0,24,111]
selection_sort(arr)
```

Insertion Sort

```python
def insertion_sort(arr):
    for i in range(1,len(arr)):
        
        for j in range(i,0,-1):
            if arr[j]<arr[j-1]:
                arr[j-1],arr[j]=arr[j],arr[j-1]
            else:
                break

    return arr

arr1=[1,3,2,78,9,2,0,24,111]
insertion_sort(arr1)
```
