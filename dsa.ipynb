LINKED LIST

```python
class Node:
    def __init__(self,data,next):
        self.data=data
        self.next=next

class LinkedList:
    def __init__(self):
        self.head=None

    def insert_at_beginning(self,data):
        if self.head is None:
            d=Node(data,None)
            self.head=d
            return
        d=Node(data,self.head)
        self.head=d
    def print(self):
        itr=self.head
        if itr is None:
            print("Nothing in this Linked List")
            return
        r=''
        while itr:
            r+=str(itr.data)+"-->"
            itr=itr.next
        print(r)
    def count(self):
        n=0
        
        itr=self.head
        if itr is None:
            print(0)
            return 0
        while itr:
            itr=itr.next
            n+=1
       
        return n
    def insert_at_end(self,data):
        if self.head is None:
            self.insert_at_beginning(data)
            return
        itr=self.head
        c=0
        count=self.count()
        while itr:
            c+=1
            if c==count:
                d=Node(data,None)
                itr.next=d
                break
            itr=itr.next
    


    def insert_values(self,list):
        self.head=None
        for data in list:
            self.insert_at_end(data)

    def insert_at_index(self,index,data):
        if index<0 or index>=self.count():
            raise Exception("Invalid index da loosuh")
        if index==0:
            self.insert_at_beginning(data)
            return
        itr=self.head
        n=0
        while itr:
            
            if n==index-1:
                d=Node(data,itr.next)
                itr.next=d
                break
            itr=itr.next
            n+=1
    def delete_at_index(self,index):
        if index<0 or index>=self.count():
            raise Exception("Invalid index da loosuh")
        if index==0:
            self.head=self.head.next
            return
        n=0
        itr=self.head
        while itr:
            
            if n==index-1:
                itr.next  =itr.next.next
                break
            itr=itr.next
            n+=1

    def insert_after_value(self,value,data):
        itr=self.head
        while itr:
            if itr.data==value:
                d=Node(data,itr.next)
                itr.next=d
                return
            itr=itr.next

        raise Exception("not found (spelling olunga poodu)")
    def remove_by_value(self,value):
        itr=self.head
        if self.head.data==value:
            self.head=self.head.next
            return
        while itr:
            
            ne=itr.next
            if ne.data==value:
                itr.next=itr.next.next
                return
            itr=itr.next
```

```python
ll=LinkedList()
ll.insert_values(["kailai","guna","ezhil","mowox","seetha"])
ll.print()
ll.insert_at_index(3,"mahesh")
ll.print()
ll.delete_at_index(0)
ll.print()
ll.insert_after_value("mahesh","kailai")
ll.print()
ll.remove_by_value("kailai")
ll.print()
```

Doubly Linkedlist

```python
class Nodes:
    def __init__(self,data,next,prev):
        self.data=data
        self.next=next
        self.prev=prev

class DoublyLL:
    def __init__(self):
        self.head=None
    def insert_at_beginning(self,data):
        if self.head is None:
            d=Nodes(data,None,None)
            self.head=d
            return
        d=Nodes(data,self.head,None)
        self.head=d
    def insert_at_end(self,data):
        if self.head is None:
            self.insert_at_beginning(data)
            return
        itr=self.head
        c=0
        count=self.count()
        while itr:
            c+=1
            if c==count:
                d=Nodes(data,None,itr)
                itr.next=d
                break
            itr=itr.next
    
    def count(self):
        n=0
        
        itr=self.head
        if itr is None:
            print(0)
            return 0
        while itr:
            itr=itr.next
            n+=1
        return n

    def insert_values(self,list):
        self.head=None
        for data in list:
            self.insert_at_end(data)
    
    def print_forward(self):
        r=''
        itr=self.head
        while itr:
            r+=str(itr.data)+"-->"
            itr=itr.next
        print(r)
    def print_backward(self):    
        r=''
        itr=self.head
        n=0
        while itr:
            if n == self.count()-1:
                rev=itr
                while rev:
                    r+="<--"+str(rev.data)
                    rev=rev.prev
            itr=itr.next
            n+=1
        print(r)
```

```python
dl=DoublyLL()
dl.insert_values(["hi","hello","bye","hell","loosuh"])
dl.print_forward()
dl.print_backward()
```

HASH MAP

```python
class HashMap:
    def __init__(self):
        self.MAX=100
        self.arr=[[] for i in range(self.MAX)]

    def hash(self,key):
        n=0
        for data in key:
            n+=ord(data)
        return n%self.MAX
    def __setitem__(self,key,value):
        n=self.hash(key)
      
        self.arr[n].append([key,value])
           
        
    def __getitem__(self,key):
        n=self.hash(key)
        for list in self.arr[n]:
            if list[0]==key:
                return list[1]
```

```python
hm=HashMap()
hm['march 29']=10
hm['may 19']=67
```
